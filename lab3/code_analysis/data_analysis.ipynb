```python
fnames = glob.glob('../data/*.npz')
#fnames = ['sun_test_140320', 'sun_test_140321', 'orion_140321', 'orion_140321_2']
#fnames = ['../data/data_'+f+'.npz' for f in fnames]
print fnames
```

Quick one-off plots

```python
for f in fnames:
    data = np.load(f)
    t = (data['jd'] - data['jd'][0])*24*3600  # Time in seconds
    v = data['volts']
    print f
    plt.figure(figsize=(6,3))
    plt.plot(t,v,'-k')
    plt.xlim([0, t[-1]])
    plt.xlabel('Time (s)', fontsize=14)
    plt.ylabel('Voltage (V)', fontsize=14)
    plt.tight_layout()
    if f == '../data/data_sun_test_140321.npz':
        print 'Figure saved'
        plt.savefig('sun_test.pdf')
    plt.show()
    
    plt.figure(figsize=(6,3))
    plt.plot(np.fft.fftfreq(v.size)[:int(v.size/2)], np.abs(np.fft.fft(v)[:int(v.size/2)])**2,'.-k')
    plt.xlim([0, 0.03])
    plt.yscale('log')
    plt.xlabel('Frequency (Hz)', fontsize=14)
    plt.ylabel('Power or something', fontsize=14)
    plt.tight_layout()
    plt.show()
```
