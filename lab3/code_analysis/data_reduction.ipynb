Data reduction functions
-------------------------

1. Obtain homing times (from log)
2. Check for errors / problems (from log)
3. Check for time gaps in time data (from .npz data)
4. Compute boxcar smoothing

We don't have to worry about the following:

* Homing always succeeds with one try
* Nothing shows up in error logs (any Exceptions would appear there)

check_log(...) was verified on moon data -- correctly identifies errors, including two instances of 2x errors (back to back)

Apply data processing

Make plots of data and stuff

This is really ugly, and I should split the data reading/reduction from the plotting.
But it works, at least.

```python
fnames = glob.glob('../data/*.npz')
lognames = glob.glob('../logs/*.txt')
print 'Files and logs:'
print fnames, '\n'
print lognames, '\n'

# FIGURES HELPED BY http://www.astrobetter.com/slim-down-your-bloated-graphics/

id_dict = {'M17': 0, 'moon': 1, 'moon_eclipse': 2, 'orion': (3,4,5), 'sun': 6, 'virgoA': 7}
log_dict = {'M17': 0, 'moon': 1, 'moon_eclipse': 2, 'orion': (3,4), 'sun': 5, 'virgoA': 6}

for name in id_dict.keys():
    # Obtain time, voltage data (merge orion data, truncate virgoA data)
    # Assign t, v, lstArr, raArr, decArr
    if name != 'orion':
        data_file = fnames[id_dict[name]]
        data = np.load(data_file)
        t = data['jd']
        v = data['volts']
        lstArr, raArr, decArr = (data['lst'], data['ra'], data['dec'])
    else:
        t = []
        v = []
        lstArr = []
        raArr = []
        decArr = []
        for idnum in id_dict['orion']:
            data_file = fnames[idnum]
            data = np.load(data_file)
            t = np.r_[t, data['jd']]
            v = np.r_[v, data['volts']]
            lstArr = np.r_[lstArr, data['lst']]
            raArr = np.r_[raArr, data['ra']]
            decArr = np.r_[decArr, data['dec']]
    
    # Obtain cut_mask (merge masks for two orion logs)
    # Also get all_cuts for plotting
    if name != 'orion':
        log_file = lognames[log_dict[name]]
        cut_mask, all_cuts = dr.get_log_cuts(t, log_file)
        if name == 'virgoA':
            # Virgo data extends too far.  Cut at last pointing
            last_pnt = 2456751.04003  # From log file, Julian day
            cut_mask = cut_mask & (t < last_pnt)
    elif name == 'orion':
        cut_mask = np.ones_like(t, dtype='bool')
        all_cuts = []
        for idnum in log_dict['orion']:
            log_file = lognames[idnum]
            curr_mask, curr_cuts = dr.get_log_cuts(t, log_file)
            cut_mask = cut_mask & curr_mask
            all_cuts = all_cuts + curr_cuts
        # Extra cut for Leo's orion data
        leo_cut = v > -0.0012
        cut_mask = cut_mask & leo_cut
    
    # Apply cuts, get boxcar smooth and subtract smoothed data
    t_cut = t[cut_mask]
    v_cut = v[cut_mask]
    v_smooth = dr.boxcar_smooth(t_cut, v_cut)
    v_shift = v_cut - v_smooth
    
    np.savez('../data_clean/'+name+'_clean.npz', volts = v_shift, jd = t_cut,
             lst = lstArr[cut_mask], ra = raArr[cut_mask], dec = decArr[cut_mask])
    
    # Apply FFT
    frq = np.fft.fftfreq(v_shift.size, 1)[:int(v_shift.size/2)]  # Assume 1 sec. spacing...
    pwr = np.abs(np.fft.fft(v_shift)[:int(v_shift.size/2)])**2
    
    print 'Plots for ' + name
    
    # Plot original data, with highlighted cut, gap locations
    plt.figure(figsize=(6,5))
    plt.subplot(3,1,1)
    for cut in all_cuts:
        plt.axvspan(cut[0], cut[1], facecolor='m', alpha=0.4)
    for gap in dr.get_gaps(t):
        plt.axvspan(gap[0], gap[1], facecolor='g', alpha=0.4)
    plt.plot(t, v, '.-b', alpha=0.2, rasterized=True, markersize=1)
    plt.plot(t_cut, v_smooth, '-r', alpha=0.7, rasterized=True, linewidth=1)
    fplot('Time (Julian days)', 'Voltage (V)')

    plt.subplot(3,1,2)
    plt.plot(t_cut, v_shift, '.-b', alpha=0.2, rasterized=True, markersize=1)
    fplot('Time (Julian days)', 'Voltage (V)')

    plt.subplot(3,1,3)
    plt.plot(frq, pwr, '.-k', rasterized=True, markersize=1)
    fplot('Frequency (Hz)', 'Power ($\mathrm{V}^2$)', [0, 0.03, plt.ylim()[0], plt.ylim()[1]], ['linear','log'])
    
    plt.tight_layout()
    plt.savefig('../plots/' + name + '_cleaning.pdf', dpi=400)
    plt.show()
```

For lunar eclipse data -- generate version with homing spikes / errors removed only, no boxcar smoothing, for public distribution.

```python
name = 'moon_eclipse'

# Obtain time, voltage data (merge orion data, truncate virgoA data)
# Assign t, v, lstArr, raArr, decArr
data_file = fnames[id_dict[name]]
data = np.load(data_file)
t = data['jd']
v = data['volts']
lstArr, raArr, decArr = (data['lst'], data['ra'], data['dec'])

# Obtain cut_mask (merge masks for two orion logs)
# Also get all_cuts for plotting
log_file = lognames[log_dict[name]]
cut_mask, all_cuts = dr.get_log_cuts(t, log_file)

# Apply cuts and save smoothed data
t_cut = t[cut_mask]
v_cut = v[cut_mask]
np.savez('../data_clean/'+name+'_dehomed.npz', volts = v_cut, jd = t_cut,
         lst = lstArr[cut_mask], ra = raArr[cut_mask], dec = decArr[cut_mask])

# Get boxcar smooth too
v_smooth = dr.boxcar_smooth(t_cut, v_cut)
v_shift = v_cut - v_smooth

# Convert times to hours
def djd2hr(t):
    return (t-2456762)*24 - 7 + 12
hr = djd2hr(t)
hr_cut = djd2hr(t_cut)

# Lunar eclipse times for reference
ptl_eclipse = np.array([22+57./60, 26 + 33./60])  # 2257 to 0233 PDT
tot_eclipse = np.array([24+6./60, 25 + 24./60])  # 0006 to 0124 PDT

# Plot original data, with highlighted cut, gap locations
# All times now in hours
print 'Plots for ' + name

# Raw data with homing cuts + lunar eclipse overlay
plt.figure(figsize=(6,2))
plt.plot(hr, v, '.-b', alpha=0.2, rasterized=True, markersize=1)
# Cuts and data gaps
for cut in all_cuts:
    cut = djd2hr(np.array(cut))
    plt.axvspan(cut[0], cut[1], facecolor='m', alpha=0.4)
# Eclipse times
plt.axvspan(ptl_eclipse[0], ptl_eclipse[1], facecolor='k', alpha=0.2)
plt.axvspan(tot_eclipse[0], tot_eclipse[1], facecolor='m', alpha=0.2)
fplot('Time (hours PDT, 2014 April 14)', 'Voltage (V)')
plt.savefig('../plots/' + name + '_nice_raw.pdf', dpi=400)
plt.show()

# Data without homing spikes, with smoothed line + lunar eclipse overlay
plt.figure(figsize=(6,2))
plt.plot(hr_cut, v_cut, '.-b', alpha=0.2, rasterized=True, markersize=1)
plt.plot(hr_cut, v_smooth, '-r', alpha=0.7, rasterized=True, linewidth=1)
# Eclipse times
plt.axvspan(ptl_eclipse[0], ptl_eclipse[1], facecolor='k', alpha=0.2)
plt.axvspan(tot_eclipse[0], tot_eclipse[1], facecolor='m', alpha=0.2)
fplot('Time (hours PDT, 2014 April 14)', 'Voltage (V)')
plt.savefig('../plots/' + name + '_nice_dehomed.pdf', dpi=400)
plt.show()

# Shifted data + lunar eclipse overlay
plt.figure(figsize=(6,2))
plt.plot(hr_cut, v_shift, '.-b', alpha=0.2, rasterized=True, markersize=1)
# Eclipse times
plt.axvspan(ptl_eclipse[0], ptl_eclipse[1], facecolor='k', alpha=0.2)
plt.axvspan(tot_eclipse[0], tot_eclipse[1], facecolor='m', alpha=0.2)
fplot('Time (hours PDT, 2014 April 14)', 'Voltage (V)')
plt.savefig('../plots/' + name + '_nice_clean.pdf', dpi=400)
plt.show()
```
