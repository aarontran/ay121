We now examine the interferometer response to extended sources in the sky, which is proportional to:
$$
    R(h_s) = \int I_\nu (\Delta h) \cos(2\pi\nu \tau_{\mathrm{tot}}) d(\Delta h)
$$
Here $h_s$ is the hour angle of the extended source's center, and $\Delta h = h - h_s$ is hour angle $h$ measured relative to our source.  This is simply the visibility equation with a cosine Fourier transform from our signal mixing, integrated over a 1-D sky.  We may think of this equation as integrating the impulse response over source extent; for $I_\nu(h_s) = \delta(\Delta h)$, we recover the usual fringe response $F(h_s) = \cos(2\pi\nu \tau_{\mathrm{tot}})$.  Conveniently, our antenna beamwidth (response) completely covers all extended sources of interest, namely the Sun and Moon.  To obtain a 1-D intensity distribution, we simply integrate source specific intensity perpendicular to the baseline direction.  For a symmetric source of small angular extent (so that we may Taylor expand in hour angle):
$$
    R(h_s) \approx \cos(2\pi\nu \tau_{\mathrm{tot}}(h_s)) \int I_\nu (\Delta h) \cos(2\pi f_f \Delta h) d(\Delta h)
$$
The left-hand $\cos$ term is simply the point source fringe response; the right-hand integral is a modulation function.  This modulation function should predict our interferometer output's envelope curve (Figure ?!?!?!?!).

The modulation function depends on the angular diameter of the source.  Consider a circular source of angular radius $R$ and uniform specific intensity; the source then has 1-D intensity profile $I_\nu(\Delta h) = \frac{1}{R} \left( R^2 - (\Delta h)^2 \right)^{1/2}$.  The corresponding modulation function is:
\begin{align*}
    \mathrm{MF} &= \int \frac{1}{R} \left( R^2 - (\Delta h)^2 \right)^{1/2} \cos(2\pi f_f \Delta h) d(\Delta h) \\
                &\propto \frac{1}{2\pi f_f R} \left(2\pi f_f R\right)
                         \int_0^\pi \sin^2 \theta \cos\left(2\pi f_f R \cos(\theta)\right) d\theta \\
                &\propto \frac{J_1 \left(2\pi f_f R \right)}{2\pi f_f R}
\end{align*}
where $J_1(z)$ is a Bessel function of the first kind [\textit{Abramowitz and Stegun}, 1964].

We obtained 1 observation of the Sun and 2 observations of the Moon (one taken during the 2014 April 14 total lunar eclipse).  The fringe modulation function should decrease in amplitude during the eclipse due to cooling of the moon's surface [e.g., \textit{Pettit and Nicholson}, 1930].  However, we first need to be able to accurately model the moon's modulation function outside of an eclipse -- this function will depend on the phase of the moon as well, but will at least have negligible thermal time dependence.

Observing the cooling of the moon during an eclipse allows measurement and characterization of the thermal properties of the moon's surface.

It only occurred to me well after the fact that it'd be good to get full moon (1 day later) and new moon data as well, to supplement the eclipse observation; our other moon observation was taken during a crescent moon ($\sim 22\%$ illumination, according to Wolfram|Alpha).  The next new moon is 2014 April 28; the next full moon is 2014 May 14.

```python
src = ephem.Sun()
src = ephem.Moon()
eclipse = True

if src.name == 'Moon' and eclipse:
    data = np.load(fnames['Moon_eclipse'])
else:
    data = np.load(fnames[src.name])

t = data['jd']
v = data['volts']
lst = data['lst']  # Hours
```

As the sun and moon move appreciably over the course of a day as seen from Earth, we use PyEphem to calculate RA, dec coordinates during our observations.  The expected fringe frequency can then be calculated using (1) time of our observations (hour angle), (2) declination during observation, (3) baseline length, and (4) observing wavelength.

(recordDVM did not correctly obtain ra/dec, hence this procedure)

```python
def get_pyephem_radec(t, obj):
    """
    Input:
        t (np.array): 1-d array of times in Julian day
        obj (ephem.FixedBody() or similar): PyEphem object
    Output:
        ra, dec arrays in radians
    """
    ra, dec = np.zeros_like(t), np.zeros_like(t)
    for i in xrange(t.size):
        obs.date = t[i] - 2415020  # PyEphem uses Dublin JD
        obj.compute(obs)
        ra[i], dec[i] = obj.ra, obj.dec  # Radians
    return ra, dec

ra, dec = get_pyephem_radec(t, src)
ha = lst - rad2hr(ra)  # Hours
ha = ((ha+12)%24) - 12  # Rescaled to fall between [-12, +12]
```

We construct an envelope curve by extracting peak points of the underlying fringe sinusoid.  Specifically, each point on our envelope curve is a maximum or minimum within a centered 30 second window.

```python
def get_envelope(t,y):
    """Extracts envelope curve from fringe data (method is specific to our interferometer).
    
    Criterion for envelope point:
        Envelope y-values are larger than all points within +/- 15 seconds
        Rationale: fastest fringe freq. is 0.027 Hz
        Period is 37 seconds, so peaks occur at closest 18.5 seconds apart.
    
    Assuming that fringes oscillate, this works best if y has no DC offset
    Otherwise it will only get half the available data
    Envelope curve is downsampled relative to original data
    
    Input:
        t (np.array): times in Julian dat
        y (np.array): data, should be fringe sinusoid with envelope
    Output:
        t_env, y_env (downsampled relative to t,y)... envelope curve
    """
    
    hw = 15. / (3600*24)  # Half-window of 15 seconds, converted to JD
    y_abs = np.abs(y)
    
    t_env = np.zeros_like(t)
    y_env = np.zeros_like(t)
    
    for i in xrange(t.size):
        lo = max(i-20, 0)  # Edge cases for indices
        hi = min(i+20, t.size)
        t_nbhd = t[lo:hi]  # Only check nearest ~40 points
        y_nbhd = y_abs[lo:hi]
        y_nbhd = y_nbhd[np.abs(t_nbhd - t[i]) < hw]  # Get nearby y-values
        if y_abs[i] >= np.amax(y_nbhd):
            t_env[i] = t[i]
            y_env[i] = y_abs[i]
    
    mask_env = np.nonzero(t_env)
    
    t_env = t_env[mask_env]
    y_env = y_env[mask_env]
    
    return t_env, y_env, mask_env


# Extract envelope curve
env_t, env_v, env_mask = get_envelope(t, v)
env_ha = ha[env_mask]
```

Finally, we define the modulating function as a function of radius and local fringe frequency (assuming uniform disk, brightness, no noise, etc).

```python
def modulating_funct(R, ffreq):
    """Fringe modulating function
    Attempt at theoretical fit, assumed 0.5 deg
    R is radius in radians
    ffreq is fringe frequency
    """
    x = np.linspace(-R,R,1000)
    fx = 1./R * np.sqrt(R**2 - x**2) * np.cos(2*np.pi*ffreq*x)
    return integrate_trap(fx, x)


def integrate_trap(ux, x):
    """Trapezoidal approx. of definite integral \int_{x[0]}^{x[-1]} ux dx = u.
    Basically same as numpy.trapz, but with less error checking / generality and stuff
    Returns scalar value of integral
    """
    dx = np.diff(x)
    u = np.sum(dx * (ux[1:] + ux[:-1]) / 2)
    return u
```

The theoretical modulation function depends on $f_f(HA)$, so we calculate $f_f = B_y / \lambda \cos(\delta) \cos(HA)$ for the HA range spanned by our observations.  As HA has units of radians in our calculation, $f_f$ has units of cycles/radian; this matches the units of object radius in our modulation function.

To get the usual units of cycles/second, take $f_f \times (2\pi / 86400)$.

```python
# Expected fringe frequencies during observation
ffreqs = baseline/wavelength * np.cos(dec) * np.cos(hr2rad(ha))
ffreqs = ffreqs[::100]  # Downsampled

# Predicted modulation function, using angular radius of 15 arcminutes (or 1/4 degree)
R_expected = deg2rad(0.25)
mf = np.array([modulating_funct(R_expected, freq) for freq in ffreqs])

# Scale appropriately -- sun radiates ~100x more strongly in radio
# These are just ad hoc correction factors
if src.name == 'Sun':
    mf *= 9
elif src.name == 'Moon':
    mf /= 2
```

Finally, let's make sure that everything looks sane.  Check that the envelope curve and predicted modulation function look reasonable.

```python
print src.name
plt.figure(figsize=(6,2))
plt.plot(ha, v, '.-b', alpha=0.1, rasterized=True, markersize=2)  # Data
plt.plot(env_ha, env_v, '-r', alpha=1)  # Envelope
plt.plot(ha[::100], np.abs(mf), '-g', linewidth=2)  # Prediction
plt.axhline(y=0, linewidth=1, color='k')

plt.xlabel('Hour angle (hr)')
plt.ylabel('Voltage (V)')
plt.axis('tight')
plt.tight_layout()
if eclipse and src.name == 'Moon':
    plt.ylim([-0.0006, 0.0006])
    plt.savefig('../plots_fitting/%s_fringes_envelopes.pdf' % 'moon_eclipse', dpi=400)
else:
    if src.name == 'Sun':
        plt.ylim([-0.01, 0.01])
    elif src.name == 'Moon':
        plt.ylim([-0.0006, 0.0006])
    plt.savefig('../plots_fitting/%s_fringes_envelopes.pdf' % src.name.lower(), dpi=400)
plt.show()
```

Modulation minima fitting
-------------------------

Once the envelope curve is obtained, we need least squares fitting routines to get minima (polynomial fit, mainly).

```python
import least_squares as lsq
```

We manually select fit domains for apparent minima in the envelope curves and perform quadratic least squares fits to the minima domains.

```python
if src.name == 'Moon':
    if eclipse:
        fit_domains = [(-4, -2.5), (2.5, 4)]
    else:
        fit_domains = [(-3.8, -2.0), (2.5, 3.8)]
elif src.name == 'Sun':
    fit_domains = [(-3.6, -3.1), (-1.5, -0.5), (0.7, 1.5), (3.3, 4.0)]

fit_coeffs = []

# Fit each domain...
for lo, hi in fit_domains:
    fit_mask = (lo < env_ha) & (env_ha < hi)
    a, s2, a_vars = lsq.poly_lsq(env_ha[fit_mask], env_v[fit_mask], n=2)
    fit_coeffs.append(a)
    fitfunc = lsq.poly(a)
    
    print '%s fit: (%g, %g)' % (src.name, lo, hi)
    plt.figure(figsize=(3,3))
    
    ax1 = plt.subplot2grid((5,1), (0,0), rowspan=3)
    plt.plot(env_ha[fit_mask], env_v[fit_mask], 'bo', markersize=5)
    plt.plot(env_ha[fit_mask], fitfunc(env_ha[fit_mask]), '-r', linewidth=3)
    plt.xlabel('Hour angle (hr.)')
    plt.ylabel('Voltage (V)')
    plt.axis('tight')
    plt.plot(ha, v, '.-g', alpha=0.5, rasterized=True, markersize=2, zorder=0)
    
    ax2 = plt.subplot2grid((5,1), (3,0), rowspan=2)
    plt.axhline(y=0, linewidth=2, color='k', linestyle='--')
    plt.plot(env_ha[fit_mask], env_v[fit_mask] - fitfunc(env_ha[fit_mask]), '.b', markersize=4)
    plt.xlabel('Hour angle (hr.)')
    plt.ylabel('Residuals (V)')
    plt.axis('tight')
    plt.ylim([-0.0001, 0.0001])
    
    plt.tight_layout()
    if eclipse and src.name == 'Moon':
        plt.savefig('../plots_fitting/%s_minimum_%g_%g.pdf' % ('moon_eclipse', lo, hi), dpi=400)
    else:
        plt.savefig('../plots_fitting/%s_minimum_%g_%g.pdf' % (src.name.lower(), lo, hi), dpi=400)
    plt.show()
```

One more plot of all fits, overlaid on the fringes.

```python
plt.figure(figsize=(6,2))
plt.plot(ha, v, '.-b', alpha=0.1, rasterized=True, markersize=2)

for coeffs, domain in zip(fit_coeffs, fit_domains):
    fit_mask = (domain[0] < env_ha) & (env_ha < domain[1])
    fitfunc = lsq.poly(coeffs)
    
    # plt.plot(env_ha[fit_mask], env_v[fit_mask], 'b.', markersize=3)
    plt.plot(env_ha, fitfunc(env_ha), '-r', linewidth=2, alpha=0.3)
    plt.plot(env_ha[fit_mask], fitfunc(env_ha[fit_mask]), '-r', linewidth=3)
    plt.tight_layout()

plt.xlabel('Hour angle (hr.)')
plt.ylabel('Voltage (V)')
plt.axis('tight')
if eclipse and src.name == 'Moon':
    plt.ylim([-0.0006, 0.0006])
else:
    if src.name == 'Sun':
        plt.ylim([-0.01, 0.01])
    elif src.name == 'Moon':
        plt.ylim([-0.0006, 0.0006])
plt.tight_layout()
if eclipse and src.name == 'Moon':
    plt.savefig('../plots_fitting/%s_minima_all.pdf' % 'moon_eclipse', dpi=400)
else:
    plt.savefig('../plots_fitting/%s_minima_all.pdf' % src.name.lower(), dpi=400)
plt.show()
```

Calculate the fitted quadratic minima, and therefore back out an estimate of angular diameter.

$f(x) = c_0 + c_1 x + c_2 x^2$ has minimum where $f'(x) = c_1 + 2 c_2 x = 0$, or $x = - c_1 / (2 c_2)$.  This gives the hour angle of our minimum, from which we get a fringe frequency at the minimum.

We expect minima in the modulation function to correspond to zero crossings of $J_1(z)$, where $z = 2\pi f_f R$.  Given an $f_f$, and the tabulated zeros of $J_1$, we obtain angular radius estimates for the sun/moon.  How do we select or identify a unique result?  Depending on our choice of radius, we will obtain some number of minima, and some shape, for the modulation function.  For the sun, we expect to see two sets of minima (reflected across HA = 0); for the moon, we expect only one set.

Although we cannot prove this (how could we make this more convincing?) -- I believe that the outside minima for the sun correspond to the 2nd zero of $J_1$; the inside minima correspond to the 3rd zero of $J_1$.  We did not fit minima at lower frequencies, as the signal was increasingly inconsistent and did not strengthen as expected.

* sun: 0.285, 0.269, 0.271, 0.319
* moon, eclipsed: 0.279, 0.272 (or, 0.152, 0.148)
* moon, not eclipsed: 0.306, 0.301 (or, 0.167, 0.164)

We could calculate zero crossings by brute force searching a range of radii -- I performed this operation and the results for radii agreed to within 0.001 degrees.  So, for simplicity, I just use the known roots of $J_1(2\pi f_f R)$ and skip the brute force rigmarole.

```python
j1_zeros = [3.8317, 7.0156, 10.1735]  # First three zeros of Bessel function of order 1

for coeffs, domain in zip(fit_coeffs, fit_domains):
    print src.name + '(%g, %g)' % domain
    
    ha_min = -1*coeffs[1] / (2*coeffs[2])  # Units of hour angle in hours...
    print 'Minimum at hour angle: %g' % ha_min
    ind_min = np.searchsorted(ha, ha_min)
    dec_min = dec[ind_min]
    ff_min = baseline/wavelength * np.cos(dec_min) * np.cos(hr2rad(ha_min))
    
    j1zeros_radii = j1_zeros/(2*np.pi*ff_min)
    
    print 'Fringe frequency at minimum (cycles/radian on sky): %g' % ff_min
    print 'Fringe frequency at minimum (cycles/second in time): %g' % (ff_min*2*np.pi/86400)
    print 'Possible angular radii from zeros of J_1 (degrees):'
    print '\t' + str(rad2deg(j1zeros_radii))
    print ''
    
    # Plot possible modulation functions
    plt.figure(figsize=(6,2))
    plt.plot(ha, v, '.-b', alpha=0.1, rasterized=True, markersize=2)
    plt.axvline(ha_min, linestyle='--', color='y', linewidth=2)
    
    if np.amax(np.abs(domain)) < 2 and src.name=='Sun':  # Minima is located betwee [-2, 2] HA
        linespecs = [':m', ':r', '-k']
    else:
        linespecs = ['-k', ':r', ':m']
    
    for rvalue, linespec in zip(j1zeros_radii, linespecs):
        mf = np.array([modulating_funct(rvalue, freq) for freq in ffreqs])
        if src.name == 'Sun':
            mf *= 9
        elif src.name == 'Moon':
            mf /= 2
        plt.plot(ha[::100], np.abs(mf), linespec, linewidth=2)
        
    plt.xlabel('Hour angle (hr.)')
    plt.ylabel('Voltage (V)')
    plt.axis('tight')
    plt.tight_layout()
    if eclipse and src.name == 'Moon':
        plt.ylim([-0.0006, 0.0006])
        plt.savefig('../plots_fitting/%s_modulation_fit_%g_%g.pdf' % ('moon_eclipse', domain[0], domain[1]), dpi=400)
    else:
        if src.name == 'Sun':
            plt.ylim([-0.01, 0.01])
        elif src.name == 'Moon':
            plt.ylim([-0.0006, 0.0006])
        plt.savefig('../plots_fitting/%s_modulation_fit_%g_%g.pdf' % (src.name.lower(), domain[0], domain[1]), dpi=400)
    plt.show()
```
