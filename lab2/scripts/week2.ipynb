## Analog mixing of analog signals (double sideband)

Import .npz file as a dict.  Keys are 'arr_0', 'arr_1', ..., 'arr_3', corresponding to mixing (1 MHz, 1.05 MHz); (1 MHz, 0.95 MHz); (100 kHz, 105 kHz); (100 kHz, 95 kHz).  We collected 1024 samples at f_samp = 10 MHz.  I don't know what the hell happenened to arr_2, arr_3 data?!  Looks like it flatlined.

I only plot the first two cases.  We get the mixing tones at 0.05, 2.05/1.95 MHz as expected.  There is a small signal near 1 MHz which could be caused by a small DC offset.  But, there is a strong signal at 4.05/3.95 MHz which I'm at a loss to explain.

Since the tone is offset by $\pm$0.05 MHz, I think it's tied to our inputs.  We could get it by mixing our 2nd signal with 3 MHz, or 5 MHz.  Or, some higher freq. signal that is subsequently aliased down.

```python
data = np.load('../data-week2/analog_mixing.npz')

fsamp = 10e6
nsamp = 1024
times = np.arange(nsamp)/fsamp * 1e6  # Microseconds
freqs = np.fft.fftfreq(1024, 1./fsamp) / 1e6  # MHz

for i in xrange(2):
    sig = data['arr_' + str(i)]
    freq_spectrum = np.fft.fft(sig)
    print 'Max power density for arr_'+str(i)+' at ' + str(freqs[np.argmax(np.abs(freq_spectrum))]) + ' MHz'
    print 'Other tone for arr_'+str(i)+' at ' + str(freqs[256+np.argmax(np.abs(freq_spectrum[256:512]))]) + ' MHz'
    
    plt.figure(figsize=(8,4))
    
    # Signal / time-domain plot
    plt.subplot(1,2,1)
    plt.plot(times, sig, '-k.')
    plt.xlabel(u'Time (\u00b5s)', fontsize=14)
    plt.ylabel(r'Voltage ($\mathrm{V}$)', fontsize=14)
    plt.text(2.5, 0.33, ['a','b'][i], weight='bold', fontsize=16)
    plt.xlim([0,50])
    if i < 2: plt.ylim([-0.4,0.4])
    #plt.yticks([-0.8, -0.4, 0, 0.4, 0.8])
    
    # Power spectrum / freq-domain plot
    plt.subplot(1,2,2)
    plt.plot(freqs[0:512], np.abs(freq_spectrum[0:512])**2,'-k*')
    plt.yscale('log')
    plt.xlabel(r'Frequency ($\mathrm{MHz}$)', fontsize=14)
    plt.ylabel(r'Power spectral density ($\mathrm{V}^2 \mathrm{Hz}^{-1}$)', fontsize=14)
    
    plt.ylim([1e-4, 1e5])
    #plt.xticks(range(0,5))
    
    plt.tight_layout()
    plt.savefig('analog_mixing_DSB_'+str(i)+'.pdf')
    plt.show()
```

Now, let's attempt to Fourier filter this signal by removing the sum frequency component (2.05, 1.95 MHz).  Then, run the inverse FFT and see if we get something sensible.

Here, I implement this by zeroing all signals with $\left|f\right| \in (1.8, 2.2) \cup (3.8, 4.2)$ MHz. 

```python
data = np.load('../data-week2/analog_mixing.npz')

fsamp = 10e6
nsamp = 1024
times = np.arange(nsamp)/fsamp * 1e6  # Microseconds
freqs = np.fft.fftfreq(1024, 1./fsamp) / 1e6  # MHz

for i in xrange(2):
    sig = data['arr_' + str(i)]
    freq_spectrum = np.fft.fft(sig)
    freq_cut = [u if np.abs(f) < 1.8 or np.abs(f) > 2.2 else 0 for (f,u) in zip(freqs, freq_spectrum)]
    print 'Max power density, after cut, for arr_'+str(i)+' at ' + str(freqs[np.argmax(np.abs(freq_cut))]) + ' MHz'
    freq_cut = [u if np.abs(f) < 3.8 or np.abs(f) > 4.2 else 0 for (f,u) in zip(freqs, freq_cut)]
    print 'Max power density, after 2nd cut, for arr_'+str(i)+' at ' + str(freqs[np.argmax(np.abs(freq_cut))]) + ' MHz'
    sig_cut = np.fft.ifft(freq_cut)  # Imaginary components are about 10^-16, as expected
    
    plt.figure(figsize=(8,4))
    
    # Signal / time-domain plot
    plt.subplot(1,2,1)
    plt.plot(times, np.real(sig_cut), '-k.')
    plt.xlabel(u'Time (\u00b5s)', fontsize=14)
    plt.ylabel(r'Voltage ($\mathrm{V}$)', fontsize=14)
    plt.xlim([0,50])
    plt.ylim([-0.06, 0.06])
    plt.text(5, 0.05, ['a','b'][i], weight='bold', fontsize=16)
    
    # Power spectrum / freq-domain plot
    plt.subplot(1,2,2)
    plt.plot(freqs[0:512], np.abs(freq_cut[0:512])**2,'-k*')
    plt.yscale('log')
    plt.xlabel(r'Frequency (MHz)', fontsize=14)
    plt.ylabel(r'Power spectral density ($\mathrm{V}^2 \mathrm{Hz}^{-1}$)', fontsize=14)
    plt.ylim([1e-4, 1e5])
    
    plt.tight_layout()
    plt.savefig('analog_mixing_DSB_filtered_'+str(i)+'.pdf')
    plt.show()
```

Introduce a small, artifial offset and see how badly it messes things up.  Here I introduce an offset of 1 time step (1/200 microseconds), which corresponds to a phase shift of 6.25/200 x 2 pi radians = 0.2 rad phase shift

```python
cos_data = np.fromfile('../data-week2/cos_bram_new', dtype='>i4')
sin_data = np.fromfile('../data-week2/sin_bram_new', dtype='>i4')

sin_data = np.roll(sin_data,-1)

fsamp = 200e6
nsamp = 2048
times = np.arange(nsamp)/fsamp * 1e6  # Microseconds
freqs = np.fft.fftfreq(2048, 1./fsamp) / 1e6  # MHz
freqs = np.roll(freqs, freqs.size/2)

sig = (cos_data + 1j*sin_data)
freq_spectrum = np.fft.fft(sig)
freq_spectrum = np.roll(freq_spectrum, freqs.size/2)

# Ad hoc way of extracting peak locations
print 'Peak at ' + str(freqs[np.argmax(np.abs(freq_spectrum[:1024]))]) + ' MHz'
print 'Peak at ' + str(freqs[1024+np.argmax(np.abs(freq_spectrum[1024:1100]))]) + ' MHz'
print 'Peak at ' + str(freqs[np.argmax(np.abs(freq_spectrum))]) + ' MHz'

plt.figure(figsize=(4,4))

# Power spectrum / freq-domain plot
plt.plot(freqs, np.abs(freq_spectrum)**2,'-ko')  # Gets rid of unsightly line
plt.yscale('log')
plt.xlabel(r'Frequency (MHz)', fontsize=14)
plt.ylabel(r'Power spectral density ($\mathrm{V}^2 \mathrm{Hz}^{-1}$)', fontsize=14)
plt.xlim([-20, 20])

# Boxes to highlight new tones, due to offset mixing
plt.axvspan(-17.5, -15, ymin=0.35, ymax=0.85, facecolor='r', alpha=0.3)
plt.axvspan(2.5, 5, ymin=0.25, ymax=0.75, facecolor='r', alpha=0.3)

plt.tight_layout()
plt.savefig('digital_mixing_SSB_offset.pdf')
plt.show()
```
