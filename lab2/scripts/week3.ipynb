## FIR filter coefficient determination and expected response

```python
nums = np.arange(-4,4)

coeffs1 = np.sinc(5./8*nums)
coeffs2 = np.sinc(5./8*nums/np.pi)  # Cutoff-freq. is now 5/(8*pi) * 100 MHz ~ 0.199 MHz
print coeffs1
print coeffs2

fsamp = 200e6
nsamp = 8
times = np.arange(nsamp)/fsamp * 1e6  # Microseconds
freqs = np.fft.fftfreq(nsamp, 1./fsamp) / 1e6  # MHz
freqs = np.roll(freqs, 4)

freq_resp_1 = np.roll(np.fft.fft(coeffs1), 4)
freq_resp_2 = np.roll(np.fft.fft(coeffs2), 4)

plt.figure(figsize=(4,3))
plt.plot(freqs, np.abs(freq_resp_1), '-ko')
plt.plot(freqs, np.abs(freq_resp_2), '-ro')
plt.xlabel(r'Frequency ($\mathrm{MHz}$)', fontsize=14)
plt.ylabel(r'Filter function (-)', fontsize=14)
plt.xlim([-100, 100])
plt.text(-85, 5, 'a', weight='bold', fontsize=16)

plt.tight_layout()
plt.savefig('FIR_filter_responses.pdf')
plt.show()
```

Filter response for many frequency bins, but only 8 coefficients (rest of FIR coefficients zeros)

```python
nums = np.arange(-4,4)
numzeros = 56

coeffs1 = np.sinc(5./8*nums)
coeffs2 = np.sinc(5./8*nums/np.pi)
coeffs1 = np.append(coeffs1, np.zeros(numzeros))
coeffs2 = np.append(coeffs2, np.zeros(numzeros))

fsamp = 200e6
nsamp = nums.size + numzeros
times = np.arange(nsamp)/fsamp * 1e6  # Microseconds
freqs = np.fft.fftfreq(nsamp, 1./fsamp) / 1e6  # MHz
freqs = np.roll(freqs, freqs.size/2)

freq_resp_1 = np.roll(np.fft.fft(coeffs1), freqs.size/2)
freq_resp_2 = np.roll(np.fft.fft(coeffs2), freqs.size/2)
freq_resp_ideal_1 = [1.6 if abs(f) < 62.5 else 0 for f in freqs]
freq_resp_ideal_2 = [1.6*np.pi if abs(f) < 5/(8*np.pi)*100 else 0 for f in freqs]

plt.figure(figsize=(4,3))
plt.plot(freqs, np.abs(freq_resp_1), ':k.')
plt.plot(freqs, np.abs(freq_resp_2), ':r.')
plt.plot(freqs, freq_resp_ideal_1, '-b.')
plt.plot(freqs, freq_resp_ideal_2, '-g.')
plt.xlabel(r'Frequency ($\mathrm{MHz}$)', fontsize=14)
plt.ylabel(r'Filter function (-)', fontsize=14)
plt.xlim([-100, 100])
plt.text(-85, 5, 'b', weight='bold', fontsize=16)

plt.tight_layout()
plt.savefig('FIR_filter_responses_fine.pdf')
plt.show()
```

## FIR filter testing (observed response)

Import all the data, then try to make sense of it.

Input signal is 30 MHz, we vary lo_freq from 10 to 60 to achieve a range of sum/difference frequencies.

```python
data_real = {}
data_imag = {}
data_both = {}
for i in [10,20,30,40,50,60]:
    data_real[i] = np.fromfile('../data-week3/ddc_real_bram_'+str(i), dtype='>i4')
    data_imag[i] = np.fromfile('../data-week3/ddc_imag_bram_'+str(i), dtype='>i4')
    data_both[i] = data_real[i] + 1j*data_imag[i]

fsamp = 200e6
nsamp = 2048
times = np.arange(nsamp)/fsamp * 1e6  # Microseconds
freqs = np.fft.fftfreq(nsamp, 1./fsamp) / 1e6  # MHz
freqs = np.roll(freqs, freqs.size/2)
```

```python
for i in [10,20,30,40,50,60]:
    sig = data_both[i]
    freq_spectrum = np.fft.fft(sig)
    freq_spectrum = np.roll(freq_spectrum, freqs.size/2)
    print 'Input frequency = 30 MHz at -30 dBm, LO frequency = ' + str(fsamp/(1e6) * i / 256) + ' MHz'
    print 'Difference frequency = ' + str( np.abs(30 - fsamp/(1e6) * i / 256) )
    print 'Sum frequency = ' + str( np.abs(30 + fsamp/(1e6) * i / 256) )
    print 'Max power density for lo_freq = '+str(i)+' at ' + str(freqs[np.argmax(np.abs(freq_spectrum))]) + ' MHz'
    
    plt.figure(figsize=(12,4))
    
    # Signal / time-domain plot
    plt.subplot(1,3,1)
    plt.plot(times, np.real(sig), '-k.')
    plt.xlabel(u'Time (\u00b5s)', fontsize=14)
    plt.ylabel(r'Voltage ($\mathrm{V}$)', fontsize=14)
    plt.xlim([0,4])
    
    # Power spectrum / freq-domain plot
    plt.subplot(1,3,2)
    plt.plot(freqs, np.abs(freq_spectrum)**2,'-k*')
    plt.yscale('log')
    plt.xlabel(r'Frequency (MHz)', fontsize=14)
    plt.ylabel(r'Power spectral density ($\mathrm{V}^2 \mathrm{Hz}^{-1}$)', fontsize=14)
    #plt.ylim([1e-4, 1e5])
    
    # Power spectrum / freq-domain plot, zoomed in
    plt.subplot(1,3,3)
    plt.plot(freqs, np.abs(freq_spectrum)**2,'-k*')
    plt.yscale('log')
    plt.xlabel(r'Frequency (MHz)', fontsize=14)
    plt.ylabel(r'Power spectral density ($\mathrm{V}^2 \mathrm{Hz}^{-1}$)', fontsize=14)
    #plt.xlim([-50, 50])
    
    plt.tight_layout()
    plt.show()
```
